```python
# ================================
# Step 2: प्रोजेक्ट अपलोड करें
# ================================
from google.colab import files
import zipfile
import os

# अपनी Kivy प्रोजेक्ट की ZIP फाइल अपलोड करें
uploaded = files.upload()

# ZIP को extract करें
for fn in uploaded.keys():
    if fn.endswith('.zip'):
        with zipfile.ZipFile(fn, 'r') as zip_ref:
            zip_ref.extractall('/content/myproject')

# अब project path सेट करें और folder बदलें
project_path = '/content/myproject'
os.chdir(project_path)
```

```python
# ================================
# Step 3: buildozer.spec बनाएँ/एडिट करें
# ================================
if not os.path.exists('buildozer.spec'):
    !buildozer init

# Unique package name और domain सेट करना
!sed -i 's/package.name = .*$/package.name = mindmatrix/' buildozer.spec
!sed -i 's/package.domain = .*$/package.domain = com.subhashsaran/' buildozer.spec
!sed -i 's/version = .*$/version = 1.0.0/' buildozer.spec
!sed -i 's|source.dir = .*|source.dir = .|' buildozer.spec
!sed -i 's|# android.permissions = INTERNET|android.permissions = INTERNET,CAMERA|' buildozer.spec
```

```python
# ================================
# Step 4: APK Build करें
# ================================
!buildozer android clean
!buildozer -v android debug
```

```python
# ================================
# Step 5: APK डाउनलोड करें
# ================================
apk_path = None
for root, dirs, files_list in os.walk('bin'):
    for file in files_list:
        if file.endswith('.apk'):
            apk_path = os.path.join(root, file)
            break

if apk_path:
    files.download(apk_path)
else:
    print("APK build नहीं हुआ। Logs देखें।")
```
